A swarmplot is a type of **categorical scatterplot** used in data visualization to display the distribution of categorical data. It is similar to a scatterplot, but instead of overlapping points, the points are adjusted so that they **do not overlap**, giving a clearer view of the distribution of the data.

### SWARM PLOT - Categorical "Scatter Plot"

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
cars = sns.load_dataset('mpg').dropna()
```

```python
cars.shape
```

```python
cars.head()
```

```python
sns.set_style('darkgrid')
```

```python
cars.cylinders.value_counts()
```

```python
# Filter down to only cars with even number of cylinders
cars = cars[cars.cylinders.isin([4,6,8])]
```

```python
cars.cylinders.value_counts()
```

```python
sns.swarmplot(cars.horsepower)
```

```python
sns.boxplot(cars.horsepower,width=0.5,showmeans=True)
```

```python
sns.scatterplot(cars.horsepower)
```

```python
sns.swarmplot(x=cars.origin,y=cars.horsepower,palette='twilight')
```

```python
sns.swarmplot(x=cars.origin,y=cars.horsepower,palette='magma',hue=cars.cylinders)
```

```python
sns.swarmplot(x=cars.origin,y=cars.horsepower,palette='magma',hue=cars.cylinders,dodge=True)
```

# Swarmplot Overlay

### With BoxPlot

```python
usa = cars[cars.origin == 'usa']
```

```python
sns.boxplot(x=usa.cylinders,y=usa.horsepower)
sns.swarmplot(x=usa.cylinders,y=usa.horsepower)
```

```python
sns.boxplot(x=usa.cylinders,y=usa.horsepower)
sns.swarmplot(x=usa.cylinders,y=usa.horsepower,color='black')
```

```python
sns.boxplot(x=usa.cylinders,y=usa.horsepower)
sns.swarmplot(x=usa.cylinders,y=usa.horsepower,color='black',alpha=0.4)
```

```python
sns.boxplot(x=usa.cylinders,y=usa.horsepower,whis = np.inf)
sns.swarmplot(x=usa.cylinders,y=usa.horsepower,color='black',alpha=0.4)
```

### With ViolinPlot

```python
sns.violinplot(x=usa.cylinders,y=usa.horsepower)
sns.swarmplot(x=usa.cylinders,y=usa.horsepower)
```

```python
sns.violinplot(x=usa.cylinders,y=usa.horsepower)
sns.swarmplot(x=usa.cylinders,y=usa.horsepower,color='white')
```

```python
sns.violinplot(x=usa.cylinders,y=usa.horsepower,scale='width')
sns.swarmplot(x=usa.cylinders,y=usa.horsepower,color='white',size=3)
```

```python
sns.violinplot(x=usa.cylinders,y=usa.horsepower,scale='width',inner=None)
sns.swarmplot(x=usa.cylinders,y=usa.horsepower,color='white',size=3)
```

```python
sns.violinplot(x=usa.cylinders,y=usa.horsepower,scale='width',inner=None,color = 'xkcd:forest green')
sns.swarmplot(x=usa.cylinders,y=usa.horsepower,color='white',size=3)
```
