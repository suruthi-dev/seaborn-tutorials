To infer from a violin plot, you can look at the following elements:


1.   Shape of the plot: The shape of the plot can tell you about the distribution of the data. If the plot is wide in the middle and narrow at the ends, the data is likely normally distributed. If the plot is skewed to one side, the data is likely skewed.

2.   Width of the plot: The width of the plot can tell you about the variability of the data. A wider plot indicates greater variability, while a narrower plot indicates less variability.


3.   Height of the plot: The height of the plot indicates the density of data at different points along the distribution. Higher sections of the plot indicate more data points in that area.


4.   Box plot elements: The box plot part of the violin plot includes the median (line in the middle of the box), the upper and lower quartiles (top and bottom of the box), and the whiskers (lines extending from the box). These elements can tell you about the central tendency and spread of the data.






*   Shape of the plot - distribution of the data
*   Width of the plot - variability of the data
*   Height of the plot - density of data at different points
*   Box plot elements - IQR,Median


```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
cars = sns.load_dataset('mpg').dropna()
```

```python
cars.shape
```

```python
cars.head()
```

```python
sns.set_style('darkgrid')
```

```python
cars.cylinders.value_counts()
```

```python
# Filter down to only cars with even number of cylinders
cars = cars[cars.cylinders.isin([4,6,8])]
```

```python
cars.cylinders.value_counts()
```

```python
sns.violinplot(x='displacement',data=cars)
```

```python
sns.violinplot(x='cylinders',y='displacement',data=cars)
```

```python
sns.violinplot(x='cylinders',y='displacement',hue='origin',data=cars)
```

# ViolinPlot Options

```python
# same violin different countries
sns.violinplot(x='cylinders',y='displacement',hue = 'origin',split=True,data=cars[cars.origin.isin(['japan','europe'])])
plt.legend(loc=2)
```

# inner = 'quartiles'

```python
sns.violinplot(x='cylinders',y='displacement',hue = 'origin',split=True,data=cars[cars.origin.isin(['japan','europe'])],inner='quartiles')
plt.legend(loc=2)
```

# scale scale_hue

Scale - The method used to scale the width of each violin. 

```python
sns.violinplot(x='cylinders',y='displacement',hue = 'origin',
               split=True,data=cars[cars.origin.isin(['japan','europe'])],
               inner='quartiles',scale='count',scale_hue=False)
plt.legend(loc=2)
```

# Styling

```python
sns.violinplot(x='origin',y='displacement',data=cars,scale='count',
               order = ['japan','europe','usa'])
```

```python
sns.violinplot(x='origin',y='displacement',data=cars,scale='count',
               order = ['japan','europe','usa'],
               linewidth=3)
```

# KDE OPTIONS

```python
sns.violinplot(x='origin',y='displacement',data=cars,scale='count',bw=0.2)
```

```python
sns.violinplot(x='origin',y='displacement',data=cars,scale='count',bw=0.6)
```
