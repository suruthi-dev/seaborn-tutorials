distplot - Distribution Plot

distplot combines - **histogram + kdeplot + rugplot**

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
cars = sns.load_dataset('mpg').dropna()
```

```python
cars.shape
```

```python
cars.head()
```

```python
sns.set_style('darkgrid')
```

```python
cars.cylinders.value_counts()
```

```python
# Filter down to only cars with even number of cylinders
cars = cars[cars.cylinders.isin([4,6,8])]
```

```python
cars.cylinders.value_counts()
```

```python
cars.weight.value_counts()
```

```python
sns.distplot(cars.weight)
```

```python
sns.distplot(cars.weight,kde=False)
```

```python
sns.distplot(cars.weight,hist=False,rug=True)
```

# Options

# Bins

```python
sns.distplot(cars.weight,bins=20)
```

```python
# customizing own bins
sns.distplot(cars.weight,bins=[1500,2000,2500,3500,5200])
```

```python
#  Fitting own distribution
from scipy.stats import norm

sns.distplot(cars.weight,fit=norm,kde=False)
```

```python
from scipy.stats import skewnorm

sns.distplot(cars.weight,fit=skewnorm,kde=False)
```

# Styling individual plots

```python
sns.distplot(cars.weight,kde=False,hist_kws={'alpha':1})

# sns.distplot(cars.weight,kde=False,hist_kws={'alpha':1,'histtype':'step'})
# sns.distplot(cars.weight,kde=False,hist_kws={'alpha':1,'histtype':'stepfilled'})
```

```python
sns.distplot(cars.weight,kde_kws ={'lw':5,'color':'black'},hist_kws={'alpha':1})
```

```python
sns.distplot(cars.weight,kde_kws = {'bw':0.8})
```

```python
sns.distplot(cars.weight,rug=True,
             kde_kws ={'lw':5,'color':'black'},
             hist_kws={'alpha':0.5,'color':'red'},
             rug_kws={'height':0.15})
```
